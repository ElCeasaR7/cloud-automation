```python
import os
import sqlite3
import json
import hashlib
import shutil
from datetime import datetime
import pandas as pd
from google.colab import userdata

# استدعاء مكتبات الـ API الجديدة والـ Embeddings
from google import genai
from google.genai import types
import chromadb

# ==========================================
# 1. إعداد مسارات المجلدات وقواعد البيانات
# ==========================================
DB_NAME = "finance_system.db"
DOCS_FOLDER = "./docs"
PROCESSED_FOLDER = "./processed"
FAILED_FOLDER = "./failed"

# ==========================================
# 2. دالة تأسيس المجلدات وقاعدة البيانات
# ==========================================
def init_system():
    # إنشاء المجلدات الثلاثة (الوارد، المعالج، الأخطاء)
    os.makedirs(DOCS_FOLDER, exist_ok=True)
    os.makedirs(PROCESSED_FOLDER, exist_ok=True)
    os.makedirs(FAILED_FOLDER, exist_ok=True)
    print("📁 تم تأسيس المجلدات الرقمية الثلاثة (docs, processed, failed) بنجاح.")

    # تهيئة قاعدة بيانات SQLite التقليدية للجداول المالية
    conn = sqlite3.connect(DB_NAME)
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS suppliers (
            id INTEGER PRIMARY KEY AUTOINCREMENT, name TEXT UNIQUE, type TEXT, created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        )
    """)
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS invoices (
            id INTEGER PRIMARY KEY AUTOINCREMENT, supplier_id INTEGER, invoice_number TEXT, invoice_hash TEXT UNIQUE,
            invoice_date TEXT, currency TEXT DEFAULT 'EGP', amount_net REAL, amount_vat REAL, amount_total REAL,
            status TEXT, file_name TEXT, file_path TEXT, ai_response TEXT, created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
            FOREIGN KEY(supplier_id) REFERENCES suppliers(id)
        )
    """)
    conn.commit()
    conn.close()
    print("🗄️ تم تهيئة وتحديث قاعدة بيانات SQLite المحاسبية.")

# ==========================================
# تشغيل دالة التأسيس فوراً لتهيئة البيئة
# ==========================================
init_system()
```

```python
# ==========================================
# 1. تهيئة عميل Gemini الجديد وقاعدة المتجهات ChromaDB
# ==========================================
client = genai.Client(api_key=userdata.get('GEMINI_API_KEY'))
chroma_client = chromadb.PersistentClient(path="./chroma_db")
invoice_collection = chroma_client.get_or_create_collection(name="purchase_invoices")

# ==========================================
# 2. بناء الـ JSON Schema الصارم لمطابقة أعمدة الشيت والنظام المصري
# ==========================================
invoice_schema = types.Schema(
    type=types.Type.OBJECT,
    properties={
        "supplier_name": types.Schema(type=types.Type.STRING, description="اسم المورد بالكامل"),
        "tax_id": types.Schema(type=types.Type.STRING, description="الرقم الضريبي للمورد إن وجد"),
        "invoice_number": types.Schema(type=types.Type.STRING, description="رقم الفاتورة"),
        "invoice_date": types.Schema(type=types.Type.STRING, description="تاريخ الفاتورة بصيغة YYYY-MM-DD"),
        "item_description": types.Schema(type=types.Type.STRING, description="اسم الصنف أو الخدمة المشتراة"),
        "receiving_note_no": types.Schema(type=types.Type.STRING, description="رقم إذن الاستلام المخزني"),
        "po_number": types.Schema(type=types.Type.STRING, description="رقم أمر الشراء"),
        "amount_net": types.Schema(type=types.Type.NUMBER, description="أصل المبلغ قبل الضرائب"),
        "amount_vat": types.Schema(type=types.Type.NUMBER, description="قيمة ضريبة القيمة المضافة 14%"),
        "wht_1": types.Schema(type=types.Type.NUMBER, description="قيمة خصم الأرباح التجارية 1%"),
        "wht_3": types.Schema(type=types.Type.NUMBER, description="قيمة خصم الأرباح التجارية 3%"),
        "amount_total": types.Schema(type=types.Type.NUMBER, description="صافي المبلغ النهائي المستحق للمورد"),
        "notes": types.Schema(type=types.Type.STRING, description="ملاحظات إضافية تم تدوينها في الحركة"),
        "ai_summary": types.Schema(type=types.Type.STRING, description="ملخص محاسبي لغوي شامل باللغة العربية للبحث الدلالي لاحقاً")
    },
    required=["supplier_name", "invoice_number", "invoice_date", "amount_net", "amount_vat", "amount_total", "ai_summary"]
)

# ==========================================
# 3. الدوال الفرعية للمحرك (الترميز، التضمين، والتحليل)
# ==========================================
def file_hash(path):
    h = hashlib.sha256()
    with open(path, "rb") as f:
        while chunk := f.read(4096): h.update(chunk)
    return h.hexdigest()

def safe_str(v, default=""): return str(v or default).strip()

def validate_date(d):
    if not d: return datetime.now().strftime("%Y-%m-%d")
    try: datetime.strptime(d, "%Y-%m-%d"); return d
    except: return datetime.now().strftime("%Y-%m-%d")

def validate_financials(net, vat, total):
    return round(net + vat, 2) == round(total, 2)

def move_file(path, folder):
    try: shutil.move(path, os.path.join(folder, os.path.basename(path)))
    except Exception as e: print("⚠️ خطأ نقل الملف:", e)

def get_embedding(text):
    try:
        response = client.models.embed_content(model="text-embedding-004", contents=text)
        return response.embeddings[0].values
    except Exception as e:
        print(f"❌ خطأ توليد المتجهات (Embedding): {e}"); return None

def handle_excel_to_json(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    try:
        if ext == ".csv":
            try:
                df = pd.read_csv(file_path, encoding="utf-8", on_bad_lines='skip')
            except UnicodeDecodeError:
                df = pd.read_csv(file_path, encoding="cp1256", on_bad_lines='skip')
        else:
            df = pd.read_excel(file_path)

        df = df.fillna("")
        return df.head(50).to_json(orient="records", force_ascii=False)
    except Exception as e:
        print(f"❌ خطأ قراءة الملف: {e}"); return None

def analyze_document(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    system_prompt = "أنت مراجع مالي خبير. استخرج البيانات بدقة متناهية وحسب الـ Schema المطلوبة، واكتب ملخصاً عربياً وافياً في حقل ai_summary."

    try:
        if ext in [".xlsx", ".xls", ".csv"]:
            sheet_data = handle_excel_to_json(file_path)
            if not sheet_data: return None
            prompt = f"حلل حركات الشيت واستخرج البيانات كقائمة سجلات:\n{sheet_data}"
            response = client.models.generate_content(
                model="gemini-2.5-flash", contents=prompt,
                config=types.GenerateContentConfig(system_instruction=system_prompt, response_mime_type="application/json", response_schema=invoice_schema, temperature=0.0)
            )
            return json.loads(response.text)
        else:
            uploaded_file = client.files.upload(file=file_path)
            response = client.models.generate_content(
                model="gemini-2.5-flash", contents=[uploaded_file, "حلل المستند."],
                config=types.GenerateContentConfig(system_instruction=system_prompt, response_mime_type="application/json", response_schema=invoice_schema, temperature=0.0)
            )
            client.files.delete(name=uploaded_file.name)
            return json.loads(response.text)
    except Exception as e:
        print(f"❌ خطأ التحليل المالي: {e}"); return None

def save_to_system(data, file_path):
    conn = sqlite3.connect(DB_NAME)
    cursor = conn.cursor()

    # التعامل مع البيانات سواء كانت فاتورة منفردة أو قائمة فواتير مستخرجة من الشيت
    records = data if isinstance(data, list) else [data]

    for record in records:
        supplier = safe_str(record.get("supplier_name"), "Unknown")
        inv_no = safe_str(record.get("invoice_number"))
        date = validate_date(record.get("invoice_date"))
        net = float(record.get("amount_net") or 0)
        vat = float(record.get("amount_vat") or 0)
        total = float(record.get("amount_total") or 0)
        ai_summary = safe_str(record.get("ai_summary"))
        file_name = os.path.basename(file_path)

        # حماية من التكرار الدفتري
        h = file_hash(file_path) + "_" + inv_no
        h_hashed = hashlib.sha256(h.encode()).hexdigest()

        cursor.execute("SELECT id FROM invoices WHERE invoice_hash=?", (h_hashed,))
        if cursor.fetchone():
            print(f"⚠️ الحركة رقم {inv_no} للمورد {supplier} مسجلة مسبقاً (مكررة).")
            continue

        status = "Validated" if validate_financials(net, vat, total) else "Pending Review"
        cursor.execute("INSERT OR IGNORE INTO suppliers(name, type) VALUES (?, ?)", (supplier, "Trading"))
        cursor.execute("SELECT id FROM suppliers WHERE name=?", (supplier,))
        sid = cursor.fetchone()[0]

        cursor.execute("""
            INSERT INTO invoices (supplier_id, invoice_number, invoice_hash, invoice_date, amount_net, amount_vat, amount_total, status, file_name, file_path, ai_response)
            VALUES (?,?,?,?,?,?,?,?,?,?,?)
        """, (sid, inv_no, h_hashed, date, net, vat, total, status, file_name, file_path, json.dumps(record, ensure_ascii=False)))

        invoice_db_id = cursor.lastrowid

        # الخطوة الأهم: توليد الـ Embedding وحفظه في قاعدة المتجهات للـ RAG
        if ai_summary:
            vector = get_embedding(ai_summary)
            if vector:
                invoice_collection.add(
                    ids=[str(invoice_db_id)],
                    embeddings=[vector],
                    documents=[ai_summary],
                    metadatas=[{"invoice_number": inv_no, "supplier_name": supplier, "invoice_date": date, "total_amount": total}]
                )

    conn.commit()
    conn.close()
    print("🧠 تم حفظ الحركات بنجاح وتوليد المتجهات الدلالية في ChromaDB!")
    return True

# ==========================================
# 4. دالة التشغيل الرئيسية الموجهة
# ==========================================
def run_rag_system():
    print(f"🔍 فحص مجلد المداخل المستندية: '{DOCS_FOLDER}'...")
    valid_files = [f for f in os.listdir(DOCS_FOLDER) if f.lower().endswith((".pdf", ".png", ".jpg", ".jpeg", ".xlsx", ".xls", ".csv"))]

    if not valid_files:
        print(f"💡 مجلد '{DOCS_FOLDER}' فارغ حالياً.")
        return

    for file_name in valid_files:
        path = os.path.join(DOCS_FOLDER, file_name)
        if not os.path.isfile(path): continue
        print(f"\n🚀 جاري معالجة: {file_name}")
        data = analyze_document(path)
        if data:
            print("📊 تفاصيل الـ JSON المحاسبي المستخرج من الذكاء الاصطناعي:")
            print(json.dumps(data, indent=2, ensure_ascii=False))
            if save_to_system(data, path):
                move_file(path, PROCESSED_FOLDER)
                print(f"📦 تم أرشفة الملف بنجاح ونقله إلى مجلد المعالجة: {PROCESSED_FOLDER}")
        else:
            move_file(path, FAILED_FOLDER)

# تشغيل المحرك الآن
run_rag_system()
```

```python
# ==========================================
# دالة توليد المتجهات (محدثة لتوافق المكتبة الجديدة)
# ==========================================
def get_embedding(text):
    try:
        # استخدام التسمية الصحيحة للموديل في المكتبة الحديثة
        response = client.models.embed_content(
            model="text-embedding-004",
            contents=text
        )
        # استخراج القيم الرقمية للمتجهات
        return response.embeddings[0].values
    except Exception as e:
        # محاولة بديلة في حال كان الحساب يتطلب بادئة معينة
        try:
            response = client.models.embed_content(
                model="models/text-embedding-004",
                contents=text
            )
            return response.embeddings[0].values
        except Exception as e_inner:
            print(f"❌ خطأ توليد المتجهات (Embedding): {e_inner}")
            return None
```

```python
import time

# ==========================================
# دالة تحليل المستند المحدثة بآلية المحاولة التلقائية (Retry)
# ==========================================
def analyze_document(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    system_prompt = "أنت مراجع مالي خبير. استخرج البيانات بدقة متناهية وحسب الـ Schema المطلوبة، واكتب ملخصاً عربياً وافياً في حقل ai_summary."

    # تحديد أقصى عدد محاولات في حال ضغط السيرفر
    max_retries = 3
    retry_delay = 5  # الانتظار 5 ثوانٍ بين المحاولة والأخرى

    for attempt in range(max_retries):
        try:
            if ext in [".xlsx", ".xls", ".csv"]:
                sheet_data = handle_excel_to_json(file_path)
                if not sheet_data: return None
                prompt = f"حلل حركات الشيت واستخرج البيانات كقائمة سجلات:\n{sheet_data}"

                response = client.models.generate_content(
                    model="gemini-2.5-flash", contents=prompt,
                    config=types.GenerateContentConfig(
                        system_instruction=system_prompt,
                        response_mime_type="application/json",
                        response_schema=invoice_schema,
                        temperature=0.0
                    )
                )
                return json.loads(response.text)
            else:
                uploaded_file = client.files.upload(file=file_path)
                response = client.models.generate_content(
                    model="gemini-2.5-flash", contents=[uploaded_file, "حلل المستند."],
                    config=types.GenerateContentConfig(
                        system_instruction=system_prompt,
                        response_mime_type="application/json",
                        response_schema=invoice_schema,
                        temperature=0.0
                    )
                )
                client.files.delete(name=uploaded_file.name)
                return json.loads(response.text)

        except Exception as e:
            # إذا كان الخطأ بسبب ضغط السيرفر 503، نحاول مجدداً
            if "503" in str(e) and attempt < max_retries - 1:
                print(f"⚠️ السيرفر مضغوط حالياً.. جاري إعادة المحاولة تلقائياً خلال {retry_delay} ثوانٍ (محاولة {attempt + 1} من {max_retries})...")
                time.sleep(retry_delay)
            else:
                print(f"❌ خطأ التحليل المالي: {e}")
                return None
```

```python
# إعادة التشغيل بعد انتهاء ضغط السيرفر المؤقت
run_rag_system()
```

```python
# ==========================================
# دالة الحفظ المحدثة (لتصفية السطور الفارغة والمجهولة)
# ==========================================
def save_to_system(data, file_path):
    conn = sqlite3.connect(DB_NAME)
    cursor = conn.cursor()

    records = data if isinstance(data, list) else [data]
    saved_count = 0

    for record in records:
        supplier = safe_str(record.get("supplier_name"))
        inv_no = safe_str(record.get("invoice_number"))
        rec_note = safe_str(record.get("receiving_note_no"))

        # 💡 فلترة محاسبية: إذا كان السطر فارغاً تماماً من البيانات الأساسية، تخطاه فوراً
        if not supplier and not inv_no and not rec_note:
            continue

        # إذا اسم المورد فارغ، نعطيه اسماً مؤقتاً مفهوماً وليس Unknown مبهم
        if not supplier:
            supplier = "مورد بدون اسم - حركة إذن رقم " + rec_note

        # تحديد رقم مرجعي فريد للحركة (إما رقم الفاتورة أو إذن الاستلام)
        reference_no = inv_no if inv_no else rec_note

        date = validate_date(record.get("invoice_date"))
        net = float(record.get("amount_net") or 0)
        vat = float(record.get("amount_vat") or 0)
        total = float(record.get("amount_total") or 0)
        ai_summary = safe_str(record.get("ai_summary"))
        file_name = os.path.basename(file_path)

        # بناء الهاش الفريد بناءً على الرقم المرجعي والمورد لمنع التكرار الحقيقي
        h = file_hash(file_path) + "_" + reference_no + "_" + supplier
        h_hashed = hashlib.sha256(h.encode()).hexdigest()

        cursor.execute("SELECT id FROM invoices WHERE invoice_hash=?", (h_hashed,))
        if cursor.fetchone():
            print(f"⚠️ الحركة رقم {reference_no} للمورد ({supplier}) مسجلة مسبقاً.")
            continue

        status = "Validated" if validate_financials(net, vat, total) else "Pending Review"
        cursor.execute("INSERT OR IGNORE INTO suppliers(name, type) VALUES (?, ?)", (supplier, "Trading"))
        cursor.execute("SELECT id FROM suppliers WHERE name=?", (supplier,))
        sid = cursor.fetchone()[0]

        cursor.execute("""
            INSERT INTO invoices (supplier_id, invoice_number, invoice_hash, invoice_date, amount_net, amount_vat, amount_total, status, file_name, file_path, ai_response)
            VALUES (?,?,?,?,?,?,?,?,?,?,?)
        """, (sid, reference_no, h_hashed, date, net, vat, total, status, file_name, file_path, json.dumps(record, ensure_ascii=False)))

        invoice_db_id = cursor.lastrowid
        saved_count += 1

        # حفظ الملخص النظيف والـ Embedding في قاعدة المتجهات
        if ai_summary:
            vector = get_embedding(ai_summary)
            if vector:
                invoice_collection.add(
                    ids=[str(invoice_db_id)],
                    embeddings=[vector],
                    documents=[ai_summary],
                    metadatas={"invoice_number": reference_no, "supplier_name": supplier, "invoice_date": date, "total_amount": total}
                )

    conn.commit()
    conn.close()

    if saved_count > 0:
        print(f"🧠 تم حفظ عدد ({saved_count}) حركات حقيقية وتوليد متجهاتها في ChromaDB بنجاح!")
        return True
    else:
        print("💡 لم يتم العثور على حركات جديدة أو حقيقية لحفظها (كل السطور فارغة أو مكررة).")
        return True
```

```python
# ==========================================
# دالة معالجة الملفات المحدثة لتناسب الفصلة المنقوطة (sep=';')
# ==========================================
def handle_excel_to_json(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    try:
        if ext == ".csv":
            try:
                # القراءة مع تحديد أن الفاصل بين الأعمدة هو الفصلة المنقوطة
                df = pd.read_csv(file_path, encoding="utf-8", sep=";", on_bad_lines='skip')
            except UnicodeDecodeError:
                # المحاولة بالترميز العربي في حال فشل الـ UTF-8
                df = pd.read_csv(file_path, encoding="cp1256", sep=";", on_bad_lines='skip')
        else:
            df = pd.read_excel(file_path)

        # تنظيف الشيت من أي قيم فارغة لضمان جودة البيانات المرسلة للموديل
        df = df.fillna("")

        # تحويل أول 50 صف مالي إلى JSON مدمج بصيغة السجلات
        return df.head(50).to_json(orient="records", force_ascii=False)
    except Exception as e:
        print(f"❌ خطأ قراءة الملف: {e}")
        return None
```

```python
# ==========================================
# دالة تحليل شيت المشتريات بنظام التدفق الخفيف (Row by Row)
# ==========================================
def analyze_document_batched(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    system_prompt = "أنت مراجع مالي خبير. استخرج البيانات بدقة متناهية وحسب الـ Schema المطلوبة، واكتب ملخصاً عربياً وافياً في حقل ai_summary."

    if ext in [".xlsx", ".xls", ".csv"]:
        sheet_data_json = handle_excel_to_json(file_path)
        if not sheet_data_json: return None

        # تحويل النص المستخرج إلى قائمة سجلات حقيقية في بايثون
        all_records = json.loads(sheet_data_json)

        # سنأخذ أول 5 صفوف فقط كمرحلة أولى للتأكد من نجاح الـ RAG والـ Embedding
        test_records = all_records[:5]
        print(f"📊 تم تجهيز ({len(test_records)}) صفوف حقيقية للمعالجة الفردية السريعة...")

        extracted_results = []

        for idx, record in enumerate(test_records):
            # إرسال صف واحد فقط في كل طلب لتفادي ضغط السيرفر تماماً
            prompt = f"حلل هذا السطر المالي واستخرج بياناته بدقة:\n{json.dumps(record, ensure_ascii=False)}"

            for attempt in range(3):
                try:
                    response = client.models.generate_content(
                        model="gemini-2.5-flash", contents=prompt,
                        config=types.GenerateContentConfig(
                            system_instruction=system_prompt,
                            response_mime_type="application/json",
                            response_schema=invoice_schema,
                            temperature=0.0
                        )
                    )
                    result_json = json.loads(response.text)
                    extracted_results.append(result_json)
                    print(f"✅ تم مراجعة واستخراج الصف رقم ({idx + 1}) بنجاح.")
                    break # نجحت المحاولة، انتقل للصف التالي
                except Exception as e:
                    if "503" in str(e) and attempt < 2:
                        print(f"⚠️ السيرفر مضغوط عند الصف {idx+1}.. إعادة محاولة سريعة...")
                        time.sleep(3)
                    else:
                        print(f"❌ تخطي الصف {idx+1} بسبب خطأ السيرفر: {e}")
                        break

        return extracted_results if extracted_results else None
    else:
        # للمستندات الأخرى كالصورة والـ PDF تظل كما هي
        return analyze_document(file_path)

# ==========================================
# تحديث دالة المحرك الرئيسي لربط التدفق الجديد
# ==========================================
def run_rag_system_v2():
    print(f"🔍 فحص مجلد المداخل المستندية: '{DOCS_FOLDER}'...")
    valid_files = [f for f in os.listdir(DOCS_FOLDER) if f.lower().endswith((".pdf", ".png", ".jpg", ".jpeg", ".xlsx", ".xls", ".csv"))]

    if not valid_files:
        print(f"💡 مجلد '{DOCS_FOLDER}' فارغ حالياً.")
        return

    for file_name in valid_files:
        path = os.path.join(DOCS_FOLDER, file_name)
        if not os.path.isfile(path): continue
        print(f"\n🚀 جاري معالجة: {file_name}")

        # استدعاء التحليل بالتدفق الخفيف
        data = analyze_document_batched(path)
        if data:
            print(f"📊 نجح استخراج البيانات! جاري التوجيه والدفع إلى الدفاتر والـ الـ ChromaDB...")
            if save_to_system(data, path):
                move_file(path, PROCESSED_FOLDER)
                print(f"📦 تم أرشفة الملف بنجاح ونقله إلى مجلد المعالجة: {PROCESSED_FOLDER}")
        else:
            print("❌ فشلت المعالجة بالكامل بسبب انقطاع السيرفر المتكرر.")
```

```python
import time

# ==========================================
# 1. تحديث دالة الـ Embedding بالطريقة القياسية الصارمة للمكتبة الجديدة
# ==========================================
def get_embedding(text):
    try:
        # استدعاء مباشر وصريح للموديل بدون أي بادئات مسارات قديمة
        response = client.models.embed_content(
            model="text-embedding-004",
            contents=text
        )
        return response.embeddings[0].values
    except Exception as e:
        print(f"❌ خطأ توليد المتجهات (Embedding): {e}")
        return None

# ==========================================
# 2. تحديث دالة التدفق لإضافة فاصل زمني (تجنب خطأ الـ 5 طلبات في الدقيقة)
# ==========================================
def analyze_document_batched(file_path):
    ext = os.path.splitext(file_path)[1].lower()
    system_prompt = "أنت مراجع مالي خبير. استخرج البيانات بدقة متناهية وحسب الـ Schema المطلوبة، واكتب ملخصاً عربياً وافياً في حقل ai_summary."

    if ext in [".xlsx", ".xls", ".csv"]:
        sheet_data_json = handle_excel_to_json(file_path)
        if not sheet_data_json: return None

        all_records = json.loads(sheet_data_json)

        # سنأخذ أول 5 صفوف مجدداً لنتأكد من استقبال الـ Embedding النظيف تماماً
        test_records = all_records[:5]
        print(f"📊 تم تجهيز ({len(test_records)}) صفوف للمعالجة مع فاصل زمني لحماية الكوتا...")

        extracted_results = []

        for idx, record in enumerate(test_records):
            prompt = f"حلل هذا السطر المالي واستخرج بياناته بدقة:\n{json.dumps(record, ensure_ascii=False)}"

            # فاصل زمني ذكي: الانتظار 12 ثانية قبل إرسال كل صف (منعاً لخطأ 429)
            if idx > 0:
                print(f"⏳ انتظار 12 ثانية لتجنب تخطي كوتا جوجل المجانية...")
                time.sleep(12)

            for attempt in range(3):
                try:
                    response = client.models.generate_content(
                        model="gemini-2.5-flash", contents=prompt,
                        config=types.GenerateContentConfig(
                            system_instruction=system_prompt,
                            response_mime_type="application/json",
                            response_schema=invoice_schema,
                            temperature=0.0
                        )
                    )
                    result_json = json.loads(response.text)
                    extracted_results.append(result_json)
                    print(f"✅ تم مراجعة واستخراج الصف رقم ({idx + 1}) بنجاح.")
                    break
                except Exception as e:
                    if "503" in str(e) and attempt < 2:
                        print(f"⚠️ السيرفر مضغوط.. إعادة محاولة خلال 5 ثوانٍ...")
                        time.sleep(5)
                    elif "429" in str(e):
                        print("⚠️ تم رصد تجاوز الكوتا.. جاري التوقف الإجباري 15 ثانية ثم المحاولة مجدداً...")
                        time.sleep(15)
                    else:
                        print(f"❌ تخطي الصف {idx+1} بسبب خطأ: {e}")
                        break

        return extracted_results if extracted_results else None
    else:
        return analyze_document(file_path)
```
